```python
# Build the list of lists representing the initial state of the crate columns
crate_columns = [[] for otters in range(9)] # the list containing lists, where each list is a column containing crates
crate_columns

for level in range(len(crates)): 
    #= (lines[5].strip()).split(',')
    #print(stacklevel, contents)
    contents = crates[len(crates)-1-level]
    print('\n')
    print(level, contents.strip())
    #columns = [contents[0:3], contents[4:7], contents[8:11], contents]
    columns = [contents[1 + i*4] for i in range(9)]
    print(columns)
    for (column, crate) in enumerate(columns):
        if (crate != ' '): # only stack real crates
            crate_columns[column].append(crate)
    print('\n\n', crate_columns)





```

```python
instructions = lines[10:] # parse the instructions
```

```python
def parse_instruction(instruction):
    inst = instruction.strip().split(' ')
    #operand = inst[0]
    num_crates = int(inst[1])
    #action = inst[2]
    origin = int(inst[3]) - 1 # omg one indexing -_-
    destination = int(inst[5]) - 1
    print('num_crates ', num_crates, ' origin ', origin, ' destination ', destination)
    return (num_crates, origin, destination)
    
```

```python
instruction = instructions[0]
```

```python
def execute_instruction(instruction):
    (num_crates,origin,destination) = parse_instruction(instruction)
    for i in range(num_crates):
        print('\n',i)
        print(crate_columns[origin])
        if (len(crate_columns[origin]) == 0):
            print("ya done fucked up\n")
            return 0
        else:
            item = crate_columns[origin].pop()
            crate_columns[destination].append(item)
            print(crate_columns[destination])
```

```python
for instruction in instructions:
    print(instruction)
    execute_instruction(instruction)
print(crate_columns)
```
